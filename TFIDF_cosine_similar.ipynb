```python
import pandas as pd
import numpy as np

#Import data from the clean file 
df = pd.read_csv('drive/Colab Notebooks/history_reading.csv')

#Print the head of the cleaned DataFrame
df.head()
```

```python
df['noidung'] = df['noidung'].fillna('')
```

```python
from sklearn.utils import shuffle
articles = df.noidung.values
titles = df.tieude.values

articles,titles = shuffle(articles,titles)
```

```python
#Create tf-idf matrix by using noidung
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(stop_words='english')

tfidf_matrix = tfidf.fit_transform(articles)

```

```python


n=20

#các bài báo có thể gợi ý
X_train_tfidf = tfidf_matrix[:-n]
X_train_title = titles[:-n]

#các bài mà người dùng đã đọc
X_test_tfidf = tfidf_matrix[-n:]
X_test_title = titles[-n:]
```

```python
#split to user and kho
print(X_test_tfidf.shape,X_train_tfidf.shape)
```

```python
def get_top_n_rec_articles(X_train_tfidf, X_train, test_article,X_train_title, n = 5):
    # calculate similarity between the corpus (i.e. the "test" data) and the user's article
    #tính sự tương tự giữa các bài báo với nội dung các bài mà user đã đọc
    similarity_scores = X_train_tfidf.dot(test_article.toarray().T)

    sorted_indicies = np.argsort(similarity_scores, axis = 0)[::-1]

    sorted_sim_scores = similarity_scores[sorted_indicies[:n]]

    # get top n corresponding document section names
    rec_title = X_train_title[sorted_indicies[:n]]

    
    # return recommendations and corresponding article meta-data
    return rec_title, sorted_sim_scores
```
